```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
WR_N = 83 # world record number of hot dogs, N

joey_rate =  WR_N / (10*60) # joey consumption rate (glizzy per second)

usain_rate = 1 / 10 # usain bolt consumption rate (glizzy per second)

def tired_factor(rate,eta,N): # joey slow down in consumption rate (he has a higher glizzy endurance, clearly)
    
    return rate * (1-eta)**N

joey_100 = 14.5 # joey 100 personal record, private communication

def joey_100_adjusted(N): # joey record to account for number of hot dogs in body -- his tummy can handle more
    
    return joey_100 + 0.1*N

WR_100 = 9.58 # usain world record, 2009

def WR_100_adjusted(N): # world record to account for number of hot dogs in body
    
    return WR_100 + 0.2*N
```

```python
N_dogs = np.arange(1,20,1)

joey_result = N_dogs / tired_factor(joey_rate,0.001,N_dogs) + joey_100_adjusted(N_dogs)
usain_result = N_dogs / tired_factor(usain_rate,0.01,N_dogs) + WR_100_adjusted(N_dogs)
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']  # Or 'Times'
plt.rcParams['mathtext.fontset'] = 'stix'  # For math symbols to match Times
plt.rcParams['mathtext.rm'] = 'Times New Roman'

fig = plt.figure(figsize=(10*2,5*4))

ax1 = fig.add_subplot(421)
ax1.tick_params(axis="both", which="major", direction="in", labelsize=15, length=7, width=1)
ax1.tick_params(axis="both", which="minor", direction="in", labelsize=15, length=2, width=1)
ax1.set_xlabel(r"Number of Glizzys",fontsize=20,labelpad=15)
ax1.set_ylabel(r"Hot Dog + 100m time [s]",fontsize=20,labelpad=15)
ax1.minorticks_on()
ax1.minorticks_on()
ax1.tick_params(axis='x', pad=10)
ax1.set_yscale('linear')

plt.plot(N_dogs,joey_result,color='black',lw=3,label='Joey')
plt.plot(N_dogs,usain_result,color='cornflowerblue',lw=3,label='Usain')

plt.legend(fontsize=15,loc='upper left')

ax1.set_yscale('log')

plt.show()
```

```python
N_dogs = np.arange(1,10000,1)

joey_result = []
usain_result = []

for i in range(100):

    log_multiplier = np.random.normal(loc=0, scale=0.5,size=2)
    multiplier = 10 ** log_multiplier

    joey_result.append(N_dogs / tired_factor(joey_rate,0.001*multiplier[0],N_dogs) + joey_100_adjusted(N_dogs))
    usain_result.append(N_dogs / tired_factor(usain_rate,0.01*multiplier[1],N_dogs) + WR_100_adjusted(N_dogs))
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']  # Or 'Times'
plt.rcParams['mathtext.fontset'] = 'stix'  # For math symbols to match Times
plt.rcParams['mathtext.rm'] = 'Times New Roman'

fig = plt.figure(figsize=(10*2,5*4))

ax1 = fig.add_subplot(421)
ax1.tick_params(axis="both", which="major", direction="in", labelsize=15, length=7, width=1)
ax1.tick_params(axis="both", which="minor", direction="in", labelsize=15, length=2, width=1)
ax1.set_xlabel(r"Number of Glizzys",fontsize=20,labelpad=15)
ax1.set_ylabel(r"Hot Dog + 100m time [s]",fontsize=20,labelpad=15)
ax1.minorticks_on()
ax1.minorticks_on()
ax1.tick_params(axis='x', pad=10)
ax1.set_yscale('linear')

for i in range(len(joey_result)):
    
    plt.plot(N_dogs,joey_result[i],color='black',lw=3,alpha=0.1)
    plt.plot(N_dogs,usain_result[i],color='cornflowerblue',lw=3,alpha=0.1)

plt.legend(fontsize=15,loc='upper left')

ax1.set_ylim(8,1000000000)
ax1.set_yscale('log')

plt.show()
```

```python
prob_list = []

for i in range(len(joey_result[0])):
    
    prob =  np.nansum(np.array(joey_result)[:,i] < np.array(usain_result)[:,i]) / len(np.array(joey_result)[:,0])
    prob_list.append(prob)
```
